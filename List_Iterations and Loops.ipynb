```
l1 = ['Shaleen','Lakshay','Ramesh','Edward'] #list 
for i in l1:
  print('Name : ',i) #Printing Values of the list without index
```

```
l1 = ['Shaleen','Lakshay','Ramesh','Edward'] #list 
for j in range(len(l1)): #Fetching Values using Index
  print('Names', l1[j],' is at index',j)
```

```
l1 = ['Shaleen','Lakshay','Ramesh','Edward']
for value,char in enumerate(l1): #Iterating list using Enumerate type 
  print(char)
```

```
l1 = ['Shaleen','Lakshay','Ramesh','Edward']
length = len(l1)
i = 0
while i<length:# Iterate list Using while loop 
  print(l1[i])
  i = i+1
```

```
l1 = ['Shaleen','Lakshay','Ramesh','Edward']
[print(i) for i in l1] #Iterate using list comprehension 
```

```
l1 = ['Shaleen','Lakshay','Ramesh','Edward']
print(*l1, sep = '\n') #Iteration without loop
```

```
l1 = ['Shaleen','Lakshay','Ramesh','Edward']
l2 = ['One','Two']
for i in l1: #implementing Nested Loops
  for j in l2:
    print(i,j)
```
