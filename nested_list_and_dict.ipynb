```
l1 = ['cat','bat','mouse']
l2 = []
l2 = [[j for j in i]for i in l1] #Making a nested list using a list
print (l2)
print(l2[2][2]) #Fetching Values using index values 
l2[0][1] = 'f'
print(l2) #Updating values using index 
l2.append('z') #Appending a value at the end of list l2
print(l2)
l2[1].insert(0,'y') #Inserting a value at index no
print(l2)
l2[1].pop(0) #Removving a value at index no
print(l2)
l2.remove('z')
print(l2) 
for k in l2:
  for t in k:
    print(t,end=' ')
```

```
d1 = {"title": "data_analyst","company":"voylla","moj" :"Aug"}
print(d1)
print(d1['title']) #Accessing value using key
d1.keys() #This will print the list of keys 
d1.values() #Prints the list of values
d1['company'] = 'Delloite' #Update the value of dict
print(d1)
d1.items() #Will provide each item of dictionary as a tuple 
for i in d1: #print values of dict
  print (i)
for i in d1:
  print(i,':',d1[i]) #print values 
```

```
people = {1: {'name': 'John', 'age': '27', 'sex': 'Male'},
          2: {'name': 'Marie', 'age': '22', 'sex': 'Female'}} #Nested List Creation
print(people)
print(people[1]['name']) #Accessing Nested dictionary item using Keys 
for i in people:
  for j in people[i]:
    print(j,':',people[i][j])
```
